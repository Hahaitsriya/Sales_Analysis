## Data Analysis

### Question 1 

```python
# Calculate the summary statistics
Sum_data=np.sum(files_df['Quantity Ordered'])
Mean_data=np.mean(files_df['Quantity Ordered'])
Standard_deviation_data=np.std(files_df['Quantity Ordered'])
Skweness_data=skew(files_df['Quantity Ordered'])
Kurtosis_data=kurtosis(files_df['Quantity Ordered'])

#Print results.
print("Sum of the total quantity ordered is " +str(Sum_data))
print("Mean of the quantity ordered is " +str(Mean_data))
print("Standard deviation of the quantity ordered is " +str(Standard_deviation_data))
print("Skweness of the quantity ordered is " +str(Skweness_data))
print("Kurtosis of the quantity ordered is " +str(Kurtosis_data))
```

### Question 2 

```python
import seaborn as sns

# Create a correlation matrix of all variable. 
correlation_data = files_df.corr()

# Plot a heatmap of the correlation matrix

sns.set_theme(style="ticks")
heatmap = sns.heatmap(data=correlation_data, annot=True,cmap="Blues", fmt='.2g')
heatmap.set_title("Correlation of all Variable.")
```

## Data Exploration 

### Question 1

```python
#Creating a new column sales from the price each and quantity ordered.
files_df['Sales']= files_df['Quantity Ordered']*files_df['Price Each']

files_df
```

```python
# group the files_df by month and calculate the sum of sales
Best_sales_monthly = files_df.groupby('Month')['Sales'].sum()

#Calculting the Highest sales with the month.
Best_month = Best_sales_monthly.sort_values(ascending=False)
print(Best_month)
Earning_of_sales = Best_sales_monthly[Best_month.index[0]]

#Printing the high sales with month.
print("The month with the best sales is", Best_month.index[0])
print("The earning in that month was", Earning_of_sales)
```

```python
# create a bar graph of sales
Best_month.plot(kind='bar')
plt.title("Monthly Sales of the Year 2019")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
# create a bar graph of sales
Best_sales_monthly.plot(kind='bar')
plt.title("Monthly Sales of the Year 2019")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

### Question 2 

```python
#Arranging the column city by the maximum product value.
print(files_df['City'].value_counts())
print("The city that has sold maximum product is " +str(files_df['City'].value_counts().index[0]))
```

### Question 3 

```python
#Grouping the Column city by the sum of sales in desending order.
Most_sold_product=files_df.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False)

#Creating a bar graph of the city relative to their sales per month.
plt.bar(Most_sold_product.index, Most_sold_product.values)
plt.title('Total sold products in the Year 2019')
plt.xlabel('Name of product')
plt.ylabel('Quantity Ordered')
plt.xticks(rotation=90)

plt.show()
```
